```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import utils as ut

show = 1
diag = ut.leer_csv("data/ARIA/5.CourseOfTreatment/_Diagnosis.csv", show).drop(columns=["DiagnosisTableName","DiagnosisId",
                                                                                     "Description","HistologyTableName",
                                                                                      "DateStamp","HisDescription"])
pri_diag = ut.leer_csv("data/ARIA/5.CourseOfTreatment/_PrmryDiagnosis.csv", show).drop(columns=["StageCriteria"])
sec_diag = ut.leer_csv("data/ARIA/5.CourseOfTreatment/_SecndryDiagnosis.csv", show)
```

```python
# ut.info_tabla(pri_diag)

pd.read_csv("data/ARIA/5.CourseOfTreatment/_Diagnosis.csv")
```

```python
pri = diag[diag.DiagnosisType=="PrmryDiagnosis"]\
.merge(pri_diag, on="DiagnosisSer", how="outer", suffixes = ("_diag", "_pri_diag")).drop(columns="DiagnosisType")

sec = diag[diag.DiagnosisType=="SecndryDiagnosis"]\
.merge(sec_diag, on="DiagnosisSer", how="outer", suffixes = ("_diag", "_sec_diag")).drop(columns="DiagnosisType")
```

```python
sec = sec.rename(columns={'DiagnosisCode': 'DiagnosisCode_sec',
                         'ObjectStatus': 'ObjectStatus_sec',
                         'HistologyCode': 'HistologyCode_sec'})
```

```python
# diag_f.to_csv("Diagnosis.csv", index=False)
pri.to_csv("data/ARIA_merged&selected/Diagnosis_pri.csv", index=False)
sec.to_csv("data/ARIA_merged&selected/Diagnosis_sec.csv", index=False)
```

# Datos y librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import utils as ut

path = "data/ARIA/5.CourseOfTreatment/"
show = 1
ps = ut.leer_csv(path+"_PlanSetup.csv", show).drop(columns=["StatusUserName","StructureSetSer",
                 "ViewingPlane","ViewingPlaneLRCorner","PlanSetupName","TransactionId","CopyOfSer","Comment",
                 "CreationUserName","PlanSetupId","StatusDate","CreationDate","ViewingPlaneULCorner","ProtocolPhaseId"])

rtp = ut.leer_csv(path+"_RTPlan.csv", show).drop(columns=["PlanUID","CreationUserName","PlanIntegrityHash",
                     'FractionPatternDigitsPerDay','FractionPattern',"PlanHashVersion","CreationDate",'TreatmentOrder'])

pres = ut.leer_csv(path + "_Prescription.csv", show).drop(columns=["TreatmentPhaseSer","CreationDate",'BolusFrequency',
        'BolusThickness',"PrescriptionName","CreationUserName","PrescriptionTemplateSer","PredecessorPrescriptionSer"])

presProp = ut.leer_csv(path + "PrescriptionProperty.csv", show)
```

# Join

```python
presProp2 = presProp[presProp.PropertyType==7][["PrescriptionSer","PropertyValue"]]\
.rename(columns={"PropertyValue": "FractionFrequency"})

rtp_ps_pres = rtp\
.merge(ps, on="PlanSetupSer", how='inner', suffixes = ("_rtp", "_ps"))\
.merge(pres, on="PrescriptionSer", how="left", suffixes = ("_rtpps", "_pres"))\
.merge(presProp2, on="PrescriptionSer", how="left")

rtp_ps_pres.shape
```
